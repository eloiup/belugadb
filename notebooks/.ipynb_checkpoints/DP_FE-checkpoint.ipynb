## Data Preprocessing and Feature Engineering

Based on the analysis made on **EDA.ipynb**, lets get our data ready for a model. <br>
This involves fixing some issues found with the data (such as NANs) and converting categorical data to numerical values that can be used by a Machine Learning model. <br>
We can also create new features (such as temporal features).

```python
import pandas as pd
import numpy as np
import holidays

pd.options.display.max_columns = 99
```

```python
train_path = '../data/train.csv'
test_path = '../data/test.csv'

train = pd.read_csv(train_path, parse_dates=['feature_0'])
test = pd.read_csv(test_path, parse_dates=['feature_0'])

df = pd.concat([train,test],0)
df.index = np.arange(len(df))

print (df.shape)
```

### Temporal Features

We wont be using month/year features, since the dataset period interval limits us to that regard.

```python
import datetime
from workalendar.america import Brazil

brazil_holidays = Brazil()
holidays = brazil_holidays.holidays(year=2017)
holidays = [values[0] for values in holidays]
```

```python
df['dow'] = df['feature_0'].map(lambda DT: DT.weekday())
df['doy'] = df['feature_0'].map(lambda DT: DT.timetuple().tm_yday)
df['dom'] = df['feature_0'].map(lambda DT: DT.day)
df['hour'] = df['feature_0'].map(lambda DT: DT.hour + DT.minute/60)

df['holiday'] = [1 if date in holidays else 0 for date in df['feature_0'].map(lambda DT: DT.date())]
df['day_after_holiday'] = [1 if date in holidays else 0 for date in df['feature_0'].map(lambda DT: DT.date() + datetime.timedelta(days=-1))]
df['day_before_holiday'] = [1 if date in holidays else 0 for date in df['feature_0'].map(lambda DT: DT.date() + datetime.timedelta(days=1))]
```

### Missing Data

Features that have missing data will have a feature created to identify that. <br>
Furtheron the missing data will be treated according to the feature value frequency distribution.

```python
feature_cols = sorted([col for col in df.columns if 'feature' in col])

features_with_nan = sorted(df[feature_cols].isnull().sum()[df[feature_cols].isnull().sum()!=0].index)

for col in features_with_nan:
    
    df['{}_NAN'.format(col)] = [1 if pd.isnull(X) else 0 for X in df[col]]
```

lets stop and recall the distributions seem on **EDA.ipynb** <br>
Those will define how we will treat each categorical and missing data. <br>
A better approach would be to evaluate different methods with a model, to empirically define which works better for each feature. However, that stray aways from the point of this exercise.

```python
plt.figure(figsize=(15,7.5))

for I in np.arange(1,13):

    plt.subplot(4,3,I)
    
    data = df['feature_{}'.format(I)].value_counts()
    
    values = data.values
    X = np.arange(len(values))
    
    plt.bar(X, values, color='blue')
    plt.bar(len(values)+1, df['feature_{}'.format(I)].isnull().sum(), color='red')
    
    plt.title('feature_{}'.format(I))
    plt.grid();
    
plt.tight_layout()
```

Features 1, 3, 6, 8, 9, 10, 11 and 12 seems like reasonable candidates for OHE, since they have a small interval of possible values. <br>
The missing data on those features will be ignored, as a label column has already been created. <br>
Luckily for us, thats all the features with missing values. <br>
Had one of the features with a more broad distribution had missing data, we would have to explore other methodologies. <br>
<br>
That leaves us with features 2, 4, 5, 7. <br>
We will use the frequency encoding for those features. <br>

It looks promising! <br>

Lets get on to building those features. 

```python
OHE_features = ['dom','dow','feature_1','feature_3','feature_6','feature_8','feature_9','feature_10','feature_11','feature_12']

for feature in OHE_features:
    
    df = pd.concat([df.drop(feature,1), pd.get_dummies(df[feature], prefix=feature)],1)
```

for the frequency features and the label encodding feature we should be careful not to leake data. <br>
we will be evaluating our models with validation and test sets, so we want to leave those outside of the feature engineering! <br>

lets first divide our datasets, than build the features. <br>
** its important to keep in mind that the test set we will be evaluated by is temporally divided. so we want to do the same with our test/validation sets. <br> **

```python
max_date = max(train['feature_0']); print ('max date:\t\t', max_date)

test_set = train.loc[train['feature_0']<=(max_date + datetime.timedelta(days=-30))]
test_date = max(test_set['feature_0']); print ('test split date:\t', test_date)

valid_set = test_set.loc[test_set['feature_0']<=(test_date + datetime.timedelta(days=-30))]
valid_date = max(valid_set['feature_0']); print ('valid split date:\t', valid_date)

train_set = valid_set.loc[valid_set['feature_0']<=(valid_date + datetime.timedelta(days=-30))]
train_date = max(train_set['feature_0']); print ('train split date:\t', train_date)
```

```python
train_set = df.loc[df['feature_0']<train_date]
valid_set = df.loc[df['feature_0'].between(train_date, valid_date)]
test_set = df.loc[df['feature_0'].between(valid_date, test_date)]

print (train_set.shape)
print (valid_set.shape)
print (test_set.shape)
```

```python
frequency_features = ['feature_2','feature_4','feature_5','feature_7']

feature_dict = {}

for feature in frequency_features:
    
    feature_dict[feature] = dict(train_set[feature].value_counts())
    
    keys = list(feature_dict[feature].keys())
    
    possible_labels = set(df[feature])
    
    for label in (possible_labels - feature_dict[feature].keys()):
        
        feature_dict[feature][label] = 1
```

```python
replace_features = ['feature_2','feature_4','feature_5','feature_7']

for feature in replace_features:
    
    train_set[feature] = train_set[feature].map(feature_dict[feature])
    valid_set[feature] = valid_set[feature].map(feature_dict[feature])
    test_set[feature] = test_set[feature].map(feature_dict[feature])
```

```python
train_set.drop(['feature_0','id'],1,inplace=True)
valid_set.drop(['feature_0','id'],1,inplace=True)
test_set.drop(['feature_0','id'],1,inplace=True)
```

```python
train_set.to_csv('../data/model_evaluation_data/train_set.csv', index=False)
valid_set.to_csv('../data/model_evaluation_data/valid_set.csv', index=False)
test_set.to_csv('../data/model_evaluation_data/test_set.csv', index=False)
```

```python
###############################################
#### feature engineering for whole dataset ####
###############################################

frequency_features = ['feature_2','feature_4','feature_5','feature_7']

feature_dict = {}

for feature in frequency_features:
    
    feature_dict[feature] = dict(train[feature].value_counts())
    
    keys = list(feature_dict[feature].keys())
    
    possible_labels = set(df[feature])
    
    for label in (possible_labels - feature_dict[feature].keys()):
        
        feature_dict[feature][label] = 1
```

```python
replace_features = ['feature_2','feature_4','feature_5','feature_7']

for feature in replace_features:
    
    df[feature] = df[feature].map(feature_dict[feature])

df.to_csv('../data/model_final_data/final.csv', index=False)
```
